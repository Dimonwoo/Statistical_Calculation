```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, binom, poisson, gamma
from scipy import stats as stats1
from scipy.stats import beta as beta_dist
```

# 1. Normal Distribution

```python
# Parameter for the Normal Distributon
mu = 0  # Parameter Mean
sigma = 1  # Parameter Standard Deviation

# Generate data points for the x-axis
x = np.linspace(start= mu-3*sigma, stop=mu+3*sigma, num=100)

# Calculate the Probability Density Function for Each Data Points
pdf = norm.pdf(x, loc=mu, scale=sigma)

# Create a plot of the normal distribution
plt.figure(figsize=(12, 5))
plt.plot(x, pdf, label="Normal Distributon Plot")
plt.title("Normal Distributon")
plt.xlabel("X-Axis")
plt.ylabel("Probability Density")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()


# @CDF- Cumulative Density Function
# Value for which you want to find the cumulative probability
z = 0.15
cdf = stats1.norm.cdf(z, loc=0, scale=1)
print(f"CDF at z = {z} is: {cdf}")
```

### Z Socore or Standard Normal Variate

```python
# Define your dataset as a NumPy array
data = np.array([10, 15, 20, 25, 30])

# Caclculate Mean and Standard variation
mean = np.mean(data)
std_dev = np.std(data)

# Calculate z-scores for each data point
z_scores = (data - mean) / std_dev

print("Original Data:", data)
print("Z-Scores:", z_scores)
```

# 2. Unifrom Distributon

```python
# Step 1: Generate random data following a uniform distribution
np.random.seed(42)  # Set a seed for reproducibility
low = 0
high = 10
num_samples = 1000
uniform_data = np.random.uniform(low, high, num_samples)

# Step 2: Create a histogram to visualize the uniform distribution
plt.figure(figsize=(8, 6))
plt.hist(uniform_data, bins=20, density=True, alpha=0.6, color='blue', edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.title('Uniform Distribution')
plt.grid(True)

# Plot the theoretical probability density function (PDF)
x = np.linspace(low, high, 100)
pdf = 1.0 / (high - low)  # Probability density function for uniform distribution
plt.plot(x, pdf, 'r-', lw=2, label='PDF (Theoretical)')

plt.legend()
plt.show()
```

# 3. Bernauli Distribution

```python
# Parameters
p = 0.3  # Probability of success

# Generate Bernoulli random variables (0 or 1)
num_samples = 1000
bernouli_data = np.random.choice([0, 1], size=num_samples, p=[1 - p, p])
bernouli_data

# Plot the histogram
plt.hist(bernouli_data, bins=[-1.5, 0.5, 1.5], density=True, alpha=0.75, edgecolor="black")
plt.xticks([0,1], ["Fialure (0)", "Success (1)"])
plt.xlabel('Outcome')
plt.ylabel('Probability')
plt.title(f'Bernoulli Distribution (p={p})')

plt.grid(True)
plt.tight_layout()
plt.show()
```

# 4. Binomial Distribution

```python
# Parameters for the binomial distribution
n = 10 # The number of trails
p = 0.5 # Probability of success on a single trial

# Generate a sample binomial distribution
x = np.arange(0, n+1) # Possible number of successes (0 to n)
pmf_values = binom.pmf(x, n, p)  # Probability mass function values
pmf_values

# Plot the probability mass function
plt.figure(figsize=(12, 6))
plt.bar(x, pmf_values, label=f"Binomial PMF (n={n}, p={p})")
plt.xlabel("Number of Success")
plt.ylabel("Probability")
plt.title("Binomial Distribution PMF")

plt.grid(True)
plt.legend(loc="upper left")
plt.tight_layout()
plt.show()
```

# 5. Poisson Distribution

```python
# Set the average rate (lambda) for the Poisson distribution
lambda_value = 3

# Generate a sample of Poisson-distributed data
sample_size = 1000
poisson_data = poisson.rvs(mu=lambda_value, size=sample_size)
poisson_data

# Create a Pandas DataFrame
df = pd.DataFrame({"Poisson_data": poisson_data})

# Plot a histogram of the Poisson data
plt.hist(df['Poisson_data'], bins=range(11), density=True, alpha=0.75, edgecolor="black")
plt.title('Poisson Distribution')
plt.xlabel('Number of Events (k)')
plt.ylabel('Probability')

plt.grid()
plt.xticks(range(11))
plt.show()
```

# 6. Beta Distribution

```python
# Define the shape parameters alpha and beta
alpha = 2
beta = 5

# Generate a range of x values
x = np.linspace(0, 1, 1000)
x
# Calculate the corresponding probability density function (PDF) values
pdf = beta_dist.pdf(x, alpha, beta)

# Create a plot
plt.plot(x, pdf, 'r-', lw=2, label=f'Beta({alpha}, {beta}) PDF')
plt.title(f'Beta Distribution (alpha={alpha}, beta={beta})')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()

plt.grid(True)
plt.tight_layout()
plt.show()
```

# 7. Gamma Distribution

```python
# Define shape and scale parameters
shape_param = 2
scale_param = 1

# Generate random data following the Gamma distribution
data = gamma.rvs(a=shape_param, scale=scale_param, size=1000)

df = pd.DataFrame({"Gamma_Data": data})

# Plot the histograme
plt.hist(df['Gamma_Data'], bins=30, density=True, alpha=0.6, color="blue", edgecolor="black", label="Histogram")

# Plot the probability density function (PDF) of the Gamma distribution
x = np.linspace(0 , 10, 1000)
pdf = gamma.pdf(x, a=shape_param, scale=scale_param)
plt.plot(x, pdf, 'r-', lw=2, label="Gamma PDF")

# Add labels and a legend
plt.title(f'Gamma Distribution (Alpha={shape_param}, Beta={scale_param})')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.legend()

plt.grid(True)
plt.show()
```
